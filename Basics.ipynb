```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from pandas.plotting import scatter_matrix

recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
```

```python
raw_data_count = recent_grads.shape[0]
print(raw_data_count)
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print(raw_data_count)
```

```python
ax1 = recent_grads.plot(x="Sample_size", y="Median", kind="scatter")
ax1.set_title("Median vs. Sample-size")
```

```python
ax2 = recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind="scatter")
ax2.set_title("Unenmployment Rate vs. Sample_size")
```

```python
ax3 = recent_grads.plot(x="Full_time", y="Median", kind="scatter")
ax3.set_title("Median vs. Full_time")
```

```python
ax4 = recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter")
ax4.set_title("Unemployment Rate vs. ShareWomen")
```

```python
ax5 = recent_grads.plot(x="Men", y="Median", kind="scatter")
ax5.set_title("Median vs. Men")
```

```python
ax6 = recent_grads.plot(x="Women", y="Median", kind="scatter")
ax6.set_title("Median vs. Women")
```

```python
ax7 = recent_grads.plot(x="Total", y="Median", kind="scatter")
ax7.set_title("Median vs. Total")
```

```python
ax8 = recent_grads.plot(x="ShareWomen", y="Median", kind="scatter")
ax8.set_title("Median vs. ShareWomen")
```

```python
ax7 = recent_grads.plot(x="Full_time", y="Median", kind="scatter")
ax7.set_title("Median vs. Full_time")
```

There seems to be a light negative correlation between popularity of master (Total) and income (Median). A strong positive correlation would create a diagonal line from the bottom left corner to the top right corner.

There is also a negative correlation between the share of women (ShareWomen) in a master and income (Median).

The total number of full-time employees and the median salary seem to have a light negative correlation.
